# Build System

## Vectorize and find cosine similarities

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['words'])
tfidf_matrix.shape
```

```python
# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construct a reverse mapping of indices and titles
indices = pd.Series(df.index, index=df['title'])
```

## Recommender function

```python
def content_recommender(title, cosine_sim=cosine_sim, df=df, indices=indices):
    
    #find the index number of the first song
    ind_num = df.query('format == 1').index.tolist()
    
    # Obtain the index of the book that matches the title
    idx = indices[title]
    print(idx)

    # Get the pairwsie similarity scores of all books and songs
    # And convert it into a list of tuples as described above
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort based on the cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the top matches (books and songs)
    sim_scores = sim_scores[1:500]

    # Get the indices
    #Loop through the top 50 matches and pull out only the songs
    title_indices = [i[0] for i in sim_scores if i[0] > ind_num[0]]
    
    #Get the top 20 songs
    title_indices = title_indices[0:20]

    # Return the top 20 most similar titles
    return df[['title', 'artist']].iloc[title_indices]
```
