**Explanation of process**

For these bands, I scraped the song lyrics myself from https://genius.com/. I used the working_webscraping.ipynb (same folder) to go through each band individually and save their song lyrics. After reading the json file saved from the webscrape, I took out the three columns I was interested in - band, song, and lyrics. That df of one band got concatenated to the df of the previous band so all the band dfs accumulated into one file.

I wanted to make sure everything was going smoothly, so I kept track of how many rows I should have after every concantenation. When finished, I renamed the column names and rearranged the df columns so they'd match the found lyric df. Then I saved it to a csv file to combine later.
